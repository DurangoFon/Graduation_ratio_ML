```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('./data/graduation_rate.csv')
df.head(5)
```

```python
df.describe()
```

```python
sns.displot(data=df, x='ACT composite score', kde = True)
plt.show()
```

```python
sns.histplot(data= df, x = 'parental level of education')
plt.xticks(rotation = 90)
plt.show()
```

```python
sns.displot(data = df, x = 'years to graduate',kde=True,bins=8)
plt.show()
```

```python
sns.displot(data=df,x = 'high school gpa',bins =18,kde=True)
plt.show()
```

```python
sns.displot(data = df, x = 'college gpa',kde=True)
plt.show()
```

```python
sns.displot(data = df, x = 'parental income',kde = True)
plt.show()
```

```python
sns.displot(data = df, x = 'SAT total score', kde = True)
plt.show()
```
